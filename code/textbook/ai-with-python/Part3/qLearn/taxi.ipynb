```python
%matplotlib widget
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as plt3

import numpy as np
import torch as tc
```

```python

fig = plt.figure()
ax =  fig.add_subplot(111, projection='3d') 

min = 1
max = 100
step = .1
s = np.arange(start=min, stop=max, step=step)

x1 = s / (s**2 - 2)

x2 = (s**2 + s - 2) / ((s**2 - 2) * (s + 1))
z = np.zeros(len(s))
for i in range(len(s)):
    z[i] = tc.linalg.vector_norm(tc.tensor(np.array([x1[i], x2[i]])), ord=2 )
print(z)
# setting title and labels 
ax.set_title("3D plot") 
ax.set_xlabel('x1-freq val') 
ax.set_ylabel('x2-freq val') 
ax.set_zlabel('X(s) p-norm') 

ax.plot(x1, x2, z)
fig.show()
```
