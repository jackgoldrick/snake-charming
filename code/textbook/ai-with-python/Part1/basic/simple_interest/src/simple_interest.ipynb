```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r  = .025
T = 50
b = np.empty(T+1)
b[0] = 10
```

```python
for t in range(T):
    b[t + 1] = ( 1 + r) * b[t]
```

```python
plt.plot(b, label = 'bank balance')
plt.xlabel("Years")
plt.ylabel("USDe6")
plt.legend()
plt.show()
```
