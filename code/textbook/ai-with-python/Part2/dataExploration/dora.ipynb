```python
data["Cases"] = [int(x.replace(",",""))  for x in data["Cases"]]

data["Date"] = pd.to_datetime(data["Date"])

data.info()
```

```python
# limit og data to columns we will use
# sort table according to attack rate
dataSlice = data[["Date", "Governorate", "Cases", "Deaths", "CFR (%)", "Attack Rate (per 1000)"]].sort_values("Attack Rate (per 1000)", ascending=False)

dsStart = dataSlice.loc[dataSlice.Date == "2017-05-22"].set_index("Governorate")
dsEnd = dataSlice.loc[dataSlice.Date == "2018-01-14"].set_index("Governorate")


print(dsStart)
print(dsEnd)
```

```python
def fix_governorates(data, fixGovs) :
    """ 
    This is our function _fix_governorates_; note that we must pass it
    two variables:
        - data: the dataframe we want to fix;
        - fixGovs: a dictionary of governorates to fix. 

    The function will do the following: 
        For a given dataframe, date list, and dictionary of Governorates
        loop through the keys in the dictionary and combine the the list
        of associated governorates into a new dataframe.

        Returns a new, fixed, dataframe.
    """

    # create can empty lsit
    newDF = []

    remove = []

    dateList = data["Date"].unique()

    # loop through each govenorate we need to fix 
    for key in fixGovs.keys():
        ds = data.loc[dataSlice.Governorate.isin(fixGovs[key])]
        newRows = {"Date": [],
                   "Cases": [],
                    "Deaths" : [],
                    "CFR (%)": [],
                    "Attack Rate (per 1000)": [] 
                    }
                   
        # Divisor for averages 
        num = len(fixGovs[key])

        # add the governorate values to the remove list
        remove.extend(fixGovs[key])

        # For each date, generate new values
        for d in dateList:
            # Data in the dataframe is stored as a Timestamp value]

            r  = ds[ds["Date"] == pd.Timestamp(d)]
            newRows["Date"].append(pd.Timestamp(d))
            newRows["Cases"].append(r.Cases.sum())
            newRows["Deaths"].append(r.Deaths.sum())
            newRows["CFR (%)"].append(r["CFR (%)"].sum()/num())
            newRows["Attack Rate (per 1000)"].append(r["Attack Rate (per 1000)"].sum()/num())


            # Create new DataFrame from combined data
            newRows = pd.DataFrame(newRows)
            # Assign the values to the key governate 
            newRows["Governorates"] = key
            # Add the new dataFrame to our list of new frames
            newDF.append(newRows)

        # Get an inverse filtered dataFrame from which we had before
        ds = dataSlice.loc[~dataSlice.Governorate.isin(remove)]
        newDF.append(ds)
        return pd.concat(newDF, ignore_index=True) 

                   
                   
                   
                   
                
```

```python
fix = { "Hadramaut": ["Moklla", "Say'on"], 
        "Al Hudaydah": ["Al Hudaydah", "Al-Hudaydah"], 
        "Al Jawf": ["Al Jawf", "Al_Jawf"], 
        "Al Maharah": ["Al Maharah", "AL Mahrah"], 
        "Marib": ["Marib", "Ma'areb"]
      }
```

```python
# Using %time to see how long it takes to run the function
%time dataSlice = fix_governorates(dataSlice, fix).sort_values("Attack Rate (per 1000)", ascending=False)

dsStart = dataSlice.loc[dataSlice.Date == "2017-05-22"].set_index("Governorate")
dsEnd = dataSlice.loc[dataSlice.Date == "2018-01-14"].set_index("Governorate")

print(dsStart)
print(dsEnd)
```
